```python
#Q5.
# Pandas provides a wide range of functions for manipulating data in a DataFrame. Here are some common functions and their usage:
data_csv = pd.read_csv("services.csv")
```

```python
# head() and tail(): To view the first or last few rows of a DataFrame
data_csv.head()
```

```python
data_csv.tail()
```

```python
#Counting the rows and columns in DataFrame using shape(). It returns the no. of rows and columns enclosed in a tuple
data_csv.shape
```

```python
#Summary of Statistics of DataFrame using describe() method.
data_csv.describe()
```

```python
# info(): To display information about the DataFrame, such as its data types and number of non-null values. For example:
data_csv.info()
```

```python
# dropna(): To remove rows with missing values. For example:
data_csv.dropna()
```

```python
# fillna(): To replace missing values with a specified value or method. For example:
data_csv.fillna(4)
```

```python
# groupby(): To group the DataFrame by one or more columns and apply a function to each group. For example
data_csv.groupby('alternate_name').mean()
```
